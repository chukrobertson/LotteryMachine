```python
# Data Processing for 2010-2019/20 archives
```

```python
import pandas as pd
import numpy as np
```

```python
# Load csv
# Only loading columns we need instead of removing them later
df12 = pd.read_csv("2012.csv",index_col=False,usecols=[0,1,2,3])
df13 = pd.read_csv("2013.csv",index_col=False,usecols=[0,1,2,3])
df14 = pd.read_csv("2014.csv",index_col=False,usecols=[0,1,2,3])
df15 = pd.read_csv("2015.csv",index_col=False,usecols=[0,1,2,3])
df16 = pd.read_csv("2016.csv",index_col=False,usecols=[0,1,2,3])
df17 = pd.read_csv("2017.csv",index_col=False,usecols=[0,1,2,3])
df18 = pd.read_csv("2018.csv",index_col=False,usecols=[0,1,2,3])
df19 = pd.read_csv("2019.csv",index_col=False,usecols=[0,1,2,3])
# dftest = pd.read_csv("2020.csv",index_col=False,usecols=[0,1,2,3])
df = pd.concat([df12,df13,df14,df15,df16,df17,df18,df19])
df
```

```python
# Converting the Date column into something useable
df['Date_Time'] = pd.to_datetime(df['Date']) 
df['Date_Str'] = df['Date_Time'].dt.strftime('%Y%m%d')
# df['Date_Num'] = pd.to_numeric(df['Date_Time'])
df['Date_Delta'] = (df['Date_Time'] - df['Date_Time'].min())  / np.timedelta64(1,'D')
df = df.sort_values(by ='Date' )
df.head()
```

```python
## Add leading zeros to the integer column in Python (creates strings)
 
df['Pick3str']=df['Pick3'].apply(lambda x: '{0:0>3}'.format(x))
df['Pick4str']=df['Pick4'].apply(lambda x: '{0:0>4}'.format(x))
print(df)
```

```python

df.info()
```

```python
# Create new columns for each Ball of each drawing
df['P3B1'] = df.Pick3str.apply(lambda x: x[0])
df['P3B2'] = df.Pick3str.apply(lambda x: x[1])
df['P3B3'] = df.Pick3str.apply(lambda x: x[2])

df['P4B1'] = df.Pick4str.apply(lambda x: x[0])
df['P4B2'] = df.Pick4str.apply(lambda x: x[1])
df['P4B3'] = df.Pick4str.apply(lambda x: x[2])
df['P4B4'] = df.Pick4str.apply(lambda x: x[3])
df.info
```

```python
# Convert string numbers to int
df.P3B1 = pd.to_numeric(df.P3B1, errors='coerce').fillna(0).astype(np.int64)
df.P3B2 = pd.to_numeric(df.P3B2, errors='coerce').fillna(0).astype(np.int64)
df.P3B3 = pd.to_numeric(df.P3B3, errors='coerce').fillna(0).astype(np.int64)

df.P4B1 = pd.to_numeric(df.P4B1, errors='coerce').fillna(0).astype(np.int64)
df.P4B2 = pd.to_numeric(df.P4B2, errors='coerce').fillna(0).astype(np.int64)
df.P4B3 = pd.to_numeric(df.P4B3, errors='coerce').fillna(0).astype(np.int64)
df.P4B4 = pd.to_numeric(df.P4B4, errors='coerce').fillna(0).astype(np.int64)
df.info()
```
