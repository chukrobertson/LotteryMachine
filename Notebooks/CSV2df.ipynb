```python
import pandas as pd
import numpy as np
```

```python
# Load csv
# Only loading columns we need instead of removing them later
df = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2020.csv",index_col=False,usecols=[0,1,2,3])
df.columns
```

```python
# See?
df
```

```python
# Rename and drop uneeded columns for individual dataframes
df.columns = ['Draw','Date','Pick3','Pick4']
# df.columns = ['Date','Draw','Pick3','Pick4','Pick5','Classic_Lotto','Kicker,''Rolling_Cash_5','Mega_Millions','Powerball','Lucky_For_Life']
# df.drop(['Pick5','Classic_Lotto','Kicker,''Rolling_Cash_5','Mega_Millions','Powerball','Lucky_For_Life'], axis=1, inplace=True)
df.columns
```

```python
# Have a look!
df
```

```python
# Encode Draw
df.Draw = df.Draw.replace({"FRI - MID": 0,
                           "FRI - EVE": 1,
                           "SAT - MID": 0,
                           "SAT - EVE": 1,
                           "SUN - MID": 0,
                           "SUN - EVE": 1,
                           "MON - MID": 0,
                           "MON - EVE": 1,
                           "TUE - MID": 0,
                           "TUE - EVE": 1,
                           "WED - MID": 0,
                           "WED - EVE": 1,
                           "THU - MID": 0,
                           "THU - EVE": 1})

df
```

```python
# Converting the Date column into something useable
df['Date'] = pd.to_datetime(df.Date)
df
```
