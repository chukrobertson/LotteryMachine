```python
# Only works for 2010pdf!
import tabula
import numpy as np
import pandas as pd
```

```python
# Path to the PDF to be read in
inPDF_Path = "../LottoData/UnprocessedData/PDFs/Cleaned_Stage_1/numbers2010_Cleaned.pdf"
# 1989 False 1990 False 1991 False 1992 False 1993 False 1994 False 1995 False 1996 False
# 1997 False 1998 False 1999 False 2000 Reads but incorrect 2001 Same 2002 Same 2003 almost useable
# 2004 Garbage 2005 Garbage 2006 Garbage 2007 Fantastic 2008 Garbage 2009 Garbage 2010
```

```python
# Reading the PDF into a list
pdf = tabula.read_pdf(inPDF_Path, encoding='utf-8', pages='all', stream=True)[0]
# Converting list to dataframe
df = pd.DataFrame(pdf)
df
```

```python
# Rename and drop uneeded columns for individual dataframes
df.columns = ['Draw','Date','Pick3','Pick4','Classic_Lotto','Rolling_Cash_5','Mega_Millions','Kicker','Powerball','Ten_OH']
df.drop(['Classic_Lotto','Rolling_Cash_5','Mega_Millions','Kicker','Powerball','Ten_OH'], axis=1, inplace=True)
df
```

```python
# Converting the Date column into something useable
df['Date'] = pd.to_datetime(df.Date)
df
```

```python
# Export our now cleaned data to csv for later processing into a larger dataset
export_csv = df.to_csv (r'../LottoData/ProcessedData/Final_Data/NewDataset_2010-2019_p3_p4/2010.csv', index = None, header=True)
```
