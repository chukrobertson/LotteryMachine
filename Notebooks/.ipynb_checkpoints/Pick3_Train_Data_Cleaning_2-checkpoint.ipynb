```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
df = []
# Load all csv into individual dataframes
# The goal is to cut the columns down to Day_Draw, Date, and Pick3, then combine all 9 into one df
# then combine all 9 into one df for further cleaning and ultimately a dataset to train a prediction model.
df2019 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2019.csv")
df2018 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2018.csv")
df2017 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2017.csv")
df2016 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2016.csv")
df2015 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2015.csv")
df2014 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2014.csv")
df2013 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2013.csv")
df2012 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2012.csv")
df2011 = pd.read_csv("../LottoData/UnprocessedData/CSVs/WinningNumberArchive_2011.csv")
# df2010 = pd.read_csv("../LottoData/ProcessedData/Processed_CSV/numbers2010.csv")
# Need to concat 2010 data
# Let's simplify our lives and combine all of these into one
df = pd.concat([df2019,df2018,df2017,df2016,df2015,df2014,df2013,df2012,df2011])
```

```python
df
```

```python
# Rename and drop uneeded columns for individual dataframes
df.columns = ['Draw','Date','Pick3','Pick4','Pick5','Classic_Lotto','Kicker','Rolling_Cash_5','Mega_Millions','Powerball','Lucky_For_Life']
df.drop(['Pick4','Pick5','Classic_Lotto','Kicker','Rolling_Cash_5','Mega_Millions','Powerball','Lucky_For_Life'], axis=1, inplace=True)
```

```python
df.Draw = df.Draw.replace({"FRI - MID": 0,
                           "FRI - EVE": 1,
                           "SAT - MID": 0,
                           "SAT - EVE": 1,
                           "SUN - MID": 0,
                           "SUN - EVE": 1,
                           "MON - MID": 0,
                           "MON - EVE": 1,
                           "TUE - MID": 0,
                           "TUE - EVE": 1,
                           "WED - MID": 0,
                           "WED - EVE": 1,
                           "THU - MID": 0,
                           "THU - EVE": 1})

df.Draw = df.Draw.replace({"FriMid": 0,
                           "FriEve": 1,
                           "SatMid": 0,
                           "SatEve": 1,
                           "SunMid": 0,
                           "SunEve": 1,
                           "MonMid": 0,
                           "MonEve": 1,
                           "TueMid": 0,
                           "TueEve": 1,
                           "WedMid": 0,
                           "WedEve": 1,
                           "ThuMid": 0,
                           "ThuEve": 1})
```

```python
df
```

```python
# # Converting the Date column into something useable
# # df['Date'] = pd.to_datetime(df.Date)
# df['Date'] = pd.to_datetime(df.Date)
df['Date'] = pd.to_datetime(df['Date'])    
df['Date_Delta'] = (df['Date'] - df['Date'].min())  / np.timedelta64(1,'D')
```

```python
df = df.sort_values(by ='Date_Delta' )
```
